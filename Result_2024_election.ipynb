```python
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = 'iframe'
import plotly.express as px
%matplotlib inline
```

```python
total_money_spentby_state = pd.DataFrame(
    merge_.groupby('State')['Amount spent (INR)'].sum().reset_index()
)
```

```python
total_money_spentby_state.head()
```

```python
fig=px.bar(total_money_spentby_state,x='State',y='Amount spent (INR)',color='Amount spent (INR)',labels={'State':'state','Amount spent (INR)':'AD amt'},
           title='Statewise amount spent on ADs')
fig.update_layout(xaxis={'categoryorder': 'total descending'},
                  xaxis_tickangle=-90,
                  width=800,
                  height=600)
fig.show()
```

```python
average_voter_turnout_by_state = pd.DataFrame(
    merge_.groupby('State')['Polled (%)'].mean().reset_index()
)
```

```python
average_voter_turnout_by_state.head()
```

```python
fig = px.bar(average_voter_turnout_by_state, 
             x='State', 
             y='Polled (%)', 
             color='Polled (%)', 
             labels={'State': 'State', 'Polled (%)': '% Poll'}, 
             title='Average Voter Turnout by State')
fig.update_layout(xaxis={'categoryorder': 'total descending'},
                  xaxis_tickangle=-90,
                  width=800,
                  height=600)
```
